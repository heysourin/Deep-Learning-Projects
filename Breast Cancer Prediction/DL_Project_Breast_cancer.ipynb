```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# separating feature and target
X = df.drop(columns='label', axis=1)
y = df['label']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

## Building Neural Network

```python
# impoting tesnoflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras

"""
By setting the seed with tf.random.set_seed(3),
you ensure that TensorFlow's random number generators produce
the same sequence of random numbers each time you run
the code with the same seed value. This is useful for debugging,
testing, and ensuring reproducibility of results across different
runs of your TensorFlow code.
"""
```

```python
X_train.shape
```

```python
# Setting up the layers of neural net
model = keras.Sequential([
                         keras.layers.Flatten(input_shape=(30,)), # converts 2D --> 1D data. you mention the feature numbers, does not contain o/p
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
#compiling the neural net
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Training
history = model.fit(X_train, y_train, validation_split=0.1, epochs=10)
# we will plot loss and accuracy value is 'history'
```

### Above you see every epoch loss dreasing, accuracy increasing. Also notice validatopn loss and accuracy.

### Standardise data. (Must be done after train test split)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
X_train_std
```

### Building NN again with standardized values

```python
# Setting up the layers of neural net
model = keras.Sequential([
                         keras.layers.Flatten(input_shape=(30,)), # converts 2D --> 1D data. you mention the feature numbers, does not contain o/p
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
#compiling the neural net
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Training
history = model.fit(X_train_std, y_train, validation_split=0.1, epochs=10)
# we will plot loss and accuracy value is 'history'
```

### Visualizing loss and accuray

```python
import matplotlib.pyplot as plt

x_values = range(1, len(history.history['accuracy']) + 1) # To make the x-axis count from 1

plt.plot(x_values, history.history['accuracy'])
plt.plot(x_values, history.history['val_accuracy'])

plt.title('Model Accuracy Plot')
plt.ylabel('accuracy')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc='lower right')

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, len(history.history['loss']) + 1) # To make the x-axis count from 1

plt.plot(x_values, history.history['loss'])
plt.plot(x_values, history.history['val_loss'])

plt.title('Model Loss Plot')
plt.ylabel('Loss')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc='upper right')

plt.show()
```

### Getting the accuracy values

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
loss, accuracy
```

### Building the predictive system

```python
#picking up a random value from csv file
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)
input_data_as_np_array = np.asarray(input_data)

#reshaping np array as we are predicting for one data point
input_data_reshaped = input_data_as_np_array.reshape(1, -1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)

print(prediction)

prediction_label = [np.argmax(prediction)]

print(prediction_label)

if(prediction_label[0] == 0):
  print('Tumor is Malignant')
else:
  print("Tumor is Benign")
```
